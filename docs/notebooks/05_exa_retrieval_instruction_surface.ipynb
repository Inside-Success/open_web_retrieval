# Plan #16: Exa Retrieval Instruction Surface

## Goal

Add one generic retrieval-instruction field to the shared `SearchQuery` contract and map it to Exa `systemPrompt`.

## Constraints

- No provider-specific field names in the boundary contract.
- Unsupported providers must fail loud, not ignore the field.
- Existing callers must remain valid.

## Success Criteria

1. `SearchQuery.retrieval_instruction` exists and is optional.
2. Exa request bodies include `systemPrompt` when the field is set.
3. Unsupported providers reject the field explicitly.
