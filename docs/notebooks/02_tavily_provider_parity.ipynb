# Tavily Provider Parity

- Status: planned
- Date: 2026-03-30
- Plan: `docs/plans/10_tavily_provider_parity.md`


## Goal

Add a real Tavily search adapter to `open_web_retrieval` without expanding the base contract or introducing a multi-provider framework abstraction.


## Inputs and Outputs

- Input: `SearchQuery(query, providers=("tavily",), top_k, recency_days, locale, domains_allow, domains_deny)`
- Output: normalized `list[SearchHit]`
- Provider-only extras stay in `SearchHit.raw_payload`


## Decisions

1. Direct `httpx` adapter, not LangChain/LlamaIndex wrappers.
2. Tavily only in this wave.
3. `OpenWebRetrievalClient(tavily_api_key=...)` is the only new wiring surface.
4. One live smoke query is required when `TAVILY_API_KEY` is available.
