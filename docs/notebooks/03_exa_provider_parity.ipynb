# Exa Provider Parity

- Status: planned
- Date: 2026-03-30
- Plan: `docs/plans/11_exa_provider_parity.md`


## Goal

Add a first-class Exa search adapter to `open_web_retrieval` with `type=deep` as the shared default.


## Contract

- Input: existing `SearchQuery`
- Output: normalized `list[SearchHit]`
- Exa extras stay in `raw_payload`


## Key decisions

1. No contract expansion in this wave.
2. `type=deep` is the adapter default.
3. `recency_days` maps to `startPublishedDate`.
4. `systemPrompt` remains deferred until a real consumer needs it.
