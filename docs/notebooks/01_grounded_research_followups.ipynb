# grounded-research Retrieval Follow-Ups

**Status:** planned  
**Repo:** `open_web_retrieval`  
**Trigger:** dense benchmark work in grounded-research exposed reusable retrieval and diagnostics follow-ups.


## Contract

**Current:**
- search/fetch/extract and Wave 0 tool-call observability are shipped
- downstream consumers still need manual inspection for domain/fallback diagnosis
- hard PDFs and blocked pages remain the main reusable retrieval pain points

**Target:**
1. explicit, testable fallback ordering
2. queryable diagnostics on provider/domain/extractor outcomes
3. consistent `trace_id` and `task` propagation across entrypoints


## Pre-Made Decisions

- improve the current providers before adding new ones
- keep `tool_call_logger` as the one shared observability contract
- keep API size small; prefer better defaults over more knobs
